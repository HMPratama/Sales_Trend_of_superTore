## **1. Histogram**

-Menampilakan distribusi data
- menampilkan frequensi (jumlah data yang muncul dari rentang nilai tertentu pada data)
- bersifat Univariate (1 variable): numerical

## a. Matplolib

```python
# Menapilkan distribusi dari data
plt.hist(data=df, x='Profit')
```

```python
# Mengatur ukuran grapik
plt.figure(figsize=(10,5))          # Figsize = (lebar, tinggi)
# Menapilkan distribusi dari data
plt.hist(data=df, 
         x='Profit',                # Kolom sumbu x
         bins=50)                   # jumlah bar--> by default 10
# Menambahkan judul dari grapik
plt.title('Frequency Distribution of Profit')

# Menambahkan lable pada sumbu X dan Y
plt.xlabel('Profit')
plt.ylabel('Frequency')


# Menampilkan legend
plt.legend(['Profit'])

# MEnampilkan grapik saja
plt.show()
```

## b.Seaborn

```python
# Menghapus outliers
df_clean=df[df['Profit'].between(-250,250)]
df_clean.head()
```

```python

# Mengatur ukuran grapik
plt.figure(figsize=(10,5))          # Figsize = (lebar, tinggi)

# Membuat grapik Histogram dengan menggunakan seaborn

sns.histplot(data=df_clean, 
             x='Profit',
             bins=50)

# Menambahkan judul dari grapik
plt.title('Frequency Distribution of Profit')

# Menyimpan Grapic
plt.savefig('visualisasi_profit.png')

# Menampilkan legend
plt.legend(['Profit'])

# MEnampilkan grapik saja
plt.show()
```

```python
df_clean['Profit'].kurtosis()
```

**Insigh:**
- distrubusi data tidak normal
- penyebaran profit paling banyak di 0-10 dollar

## **3. Lineplot**

- Menampikan tren dalam rentang waktu tertentu
- Menggunakan data timeseries (varible yang berkaitan dengan waktu)
- Bersifat Bivariate (2 variable): time series vs numerical

```python
df_clean['Order Date'].value_counts()
```

```python
# Mengambil nilai bulan dan tahun pada kolom order date
df_clean['Order Month']=df_clean['Order Date'].apply(lambda x: str(x)[0:7])

# Mengubah kolom Order month menjadi datetime
df_clean['Order Month']=pd.to_datetime(df_clean['Order Month'])
df_clean
```

```python
for dt in df_clean['Order Month']:
    print(dt.strftime('%m-%y'))
```

```python
# Mengatur ukuran dari Grapik
plt.figure(figsize=(10,5))

# Membuat lineplot

sns.lineplot(data=df_clean,
             x='Order Month',
             y= 'Sales',
             ci=0)

# Mengatur Posisi Xticks
plt.xticks(ticks=df_clean['Order Month'],
           labels= [dt.strftime ('%m-%y') for dt in df_clean['Order Month']],
           rotation = 90)

plt.show
```

Insight:
- Terjadi fluktuatif sales sepanjan tahun 2016 sampai 2019
- Setiap Bulan november, penjualan selalu mengalami kenaikan
- Setiap bulan Febuary sales selalu penurunan

```python
# Mengatur ukuran dari Grapik
plt.figure(figsize=(10,5))

# Membuat lineplot

sns.lineplot(data=df_clean,
             x='Order Month',
             y= 'Sales',
             ci=0,
               estimator=np.sum)

# Mengatur Posisi Xticks
plt.xticks(ticks=df_clean['Order Month'],
           labels= [dt.strftime ('%m-%y') for dt in df_clean['Order Month']],
           rotation = 90)

plt.show
```

```python
# Mengambil nilai bulan dan tahun pada kolom order date
df_clean['Order Year']=df_clean['Order Month'].apply(lambda x: str(x)[0:4])

# Mengubah kolom Order month menjadi datetime
df_clean['Order Year']=pd.to_datetime(df_clean['Order Year'])
df_clean
```

```python
# Mengatur ukuran dari Grapik
plt.figure(figsize=(10,5))

# Membuat lineplot

sns.lineplot(data=df_clean,
             x='Order Year',
             y= 'Sales',
             ci=0,
               estimator=np.sum)

# Mengatur Posisi Xticks
plt.xticks(ticks=df_clean['Order Year'],
           labels= [dt.strftime ('%Y') for dt in df_clean['Order Year']],
           rotation = 15)

plt.show()
```

Insight:
- Penjualan mengalami kenaikan dari tahun 2016 sampai tahun 2019.
- Pada 2017 mengalami kenaikan penjualan yang signifikan sampai tahun 2019

```python
df_clean['Order Year'] = df_clean['Order Year'].apply(lambda x: str(x)[0:4])
dft = df_clean.groupby('Order Year')[['Profit','Sales']].sum()

with sns.axes_style("darkgrid"):
    plt.figure(figsize=(15,10))
    ax = sns.lineplot(data = dft, x = dft.index.values, y = 'Sales', color = 'salmon',marker='*', markersize=25)
    ax.set_xticklabels(dft.index.values, rotation=45)
    plt.title('Sum of Sales by Year', size = 15)
    plt.xlabel('Year of Sales', fontsize=12)
    plt.ylabel('Sum of Sales', fontsize=12)
    plt.show()
```
